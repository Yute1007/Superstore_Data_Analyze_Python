```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Superstore.csv")
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Rename Columns
df_new = df.rename(columns = {'Row ID' : 'Row_ID', 'Order ID' : 'Order_ID', 'Order Date' : 'Order_Date','Ship Date': 'Ship_Date', 'Ship Mode': 'Ship_Mode', 'Customer ID': 'Customer_ID', 'Customer Name' : 'Customer_Name', 'Country/Region' : 'Country', 'State/Province' : 'Province', 'Sub-Category' : 'Sub_Category'}, inplace = False)
df_new
```

```python
＃I only want to have the superstore dataset in the U.S.
```

```python
df_USA = df_new[df_new['Country'] == 'United States']
df_USA
```

```python
# to know what kind of Subcategory we have
df_USA.Sub_Category.unique()
```

```python
# to know what kind of Category we have
df_USA.Category.unique()
```

```python
# What sub-category belong which Category
df_USA.groupby('Category')['Sub_Category'].apply(list)
```

```python
# Medain sales in Sub_Category
df_USA.groupby('Sub_Category')['Sales'].median().sort_values(ascending=False) 
```

```python
# Mean sales in Sub_Category
df_USA.groupby('Sub_Category')['Sales'].mean().sort_values(ascending=False) 
```

```python
# Covert the Ship_Date to right format
df_USA.loc[:, 'Ship_Date'] = pd.to_datetime(df_USA['Ship_Date'], errors='coerce')
```

```python
# Find the Sub_Category's Sales in 2023
df_USA_2023 = df_USA[df_USA['Ship_Date'].dt.year == 2023].groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).reset_index()
df_USA_2023
```

```python
# Find the Sub_Category's Sales in 2022
df_USA_2022 = df_USA[df_USA['Ship_Date'].dt.year == 2022].groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).reset_index()
df_USA_2022
```

```python
# create a bar chart to visualize the trade of Sub_Category in 2023 and 2022
df_sales_comparison = pd.merge(df_USA_2023, df_USA_2022, on='Sub_Category', suffixes=('_2023', '_2022'))
df_sales_comparison.plot(kind='bar', x='Sub_Category', y=['Sales_2023', 'Sales_2022'], 
                         title='Sub_Category Sales Comparison: 2023 vs 2022', xlabel='Sub_Category', ylabel='Sales', 
                         figsize=(10,6))
plt.show()
```

```python
# to know what kind of regions we have in this dataset
df_USA.Region.unique()
```

```python
# Average sales groupby Region
df_USA.groupby('Region')['Sales'].sum().sort_values(ascending=False) 
```

```python
df_USA_Region = df_USA.groupby('Region')['Sales'].sum().sort_values(ascending=False)
mylabels = df_USA_Region.index.tolist()
explode = (0, 0.1, 0, 0)
df_USA_Region.plot(kind='pie', title='Sum of Sales in Region', labels = mylabels, autopct='%1.1f%%',explode=explode)
```

```python
# Finding which region has biggest nagative profit 
df_USA.groupby('Region')['Profit'].min().sort_values()
```

```python
# Finding which region has biggest profit
df_USA.groupby('Region')['Profit'].max().sort_values(ascending=False)
```

```python
# what kinds of ship mode we have 
df_USA.Ship_Mode.unique()
```

```python
# which ship mode are we frequently used in 2023
df_USA[df_USA['Ship_Date'].dt.year == 2023]['Ship_Mode'].value_counts()
```

```python
# which ship mode are we frequently used in 2022
df_USA[df_USA['Ship_Date'].dt.year == 2022]['Ship_Mode'].value_counts()
```

```python
df_2023 = df_USA[df_USA['Ship_Date'].dt.year == 2023]
ship_mode_sales_2023 = df_2023.groupby('Ship_Mode')['Sales'].mean().sort_values(ascending=False)
ship_mode_sales_2023
```

```python
# Create a chart to describe the relationship between ship mode in 2023 and sales in 2023
ship_mode_sales_2023.plot(kind='bar', title='Average Sales per Ship Mode in 2023', xlabel='Ship Mode', ylabel='Average Sales', figsize=(8, 5))
plt.show()
```
